```python
#Script written to compare VAF's between two VCF files
#by: reykajayasinghe@gmail.com
#Edited on: 04/09/2019

#V1="/Volumes/Angie's Bac/Met Seq Data/Index35_OtB3992_Angela.Hirbe_900-1985a_human_hEx_AV5_30x_07222016.deduplicated.realigned.recalibrated.snpEff.snpSift.vcf"
#V2="/Volumes/Angie's Bac/Met Seq Data/Index36_OtB3993_Angela.Hirbe_900-1985a_human_hEx_AV5_30x_FFPE_07222016.deduplicated.realigned.recalibrated.snpEff.snpSift.vcf"
#V3="/Volumes/Angie's Bac/Met Seq Data/Index37_OtB3994_Angela.Hirbe_900-1985a_human_hEx_AV5_30x_FFPE_07222016.deduplicated.realigned.recalibrated.snpEff.snpSift.vcf"
#V4="/Volumes/Angie's Bac/Met Seq Data/Index38_OtB3995_Angela.Hirbe_900-1985a_human_hEx_AV5_30x_FFPE_07222016.deduplicated.realigned.recalibrated.snpEff.snpSift.vcf"

V1="/Volumes/Angie's Bac/Met Seq Data/Index39_OtB3996_Angela.Hirbe_900-1985a_human_hEx_AV5_30x_07222016.deduplicated.realigned.recalibrated.snpEff.snpSift.vcf"
V2="/Volumes/Angie's Bac/Met Seq Data/Index40_OtB3997_Angela.Hirbe_900-1985a_human_hEx_AV5_30x_FFPE_07222016.deduplicated.realigned.recalibrated.snpEff.snpSift.vcf"
V3="/Volumes/Angie's Bac/Met Seq Data/Index42_OtB3998_Angela.Hirbe_900-1985a_human_hEx_AV5_30x_FFPE_07222016.deduplicated.realigned.recalibrated.snpEff.snpSift.vcf"
V4="/Volumes/Angie's Bac/Met Seq Data/Index43_OtB3999_Angela.Hirbe_900-1985a_human_hEx_AV5_30x_FFPE_07222016.deduplicated.realigned.recalibrated.snpEff.snpSift.vcf"

#Index35_OtB3992_Angela.Hirbe_900-1985a_human_hEx_AV5_30x_07222016.deduplicated.realigned.recalibrated.snpEff.snpSift.vcf
#Index36_OtB3993_Angela.Hirbe_900-1985a_human_hEx_AV5_30x_FFPE_07222016.deduplicated.realigned.recalibrated.snpEff.snpSift.vcf
#Index37_OtB3994_Angela.Hirbe_900-1985a_human_hEx_AV5_30x_FFPE_07222016.deduplicated.realigned.recalibrated.snpEff.snpSift.vcf
#Index38_OtB3995_Angela.Hirbe_900-1985a_human_hEx_AV5_30x_FFPE_07222016.deduplicated.realigned.recalibrated.snpEff.snpSift.vcf
#Index39_OtB3996_Angela.Hirbe_900-1985a_human_hEx_AV5_30x_07222016.deduplicated.realigned.recalibrated.snpEff.snpSift.vcf
#Index40_OtB3997_Angela.Hirbe_900-1985a_human_hEx_AV5_30x_FFPE_07222016.deduplicated.realigned.recalibrated.snpEff.snpSift.vcf
#Index42_OtB3998_Angela.Hirbe_900-1985a_human_hEx_AV5_30x_FFPE_07222016.deduplicated.realigned.recalibrated.snpEff.snpSift.vcf
#Index43_OtB3999_Angela.Hirbe_900-1985a_human_hEx_AV5_30x_FFPE_07222016.deduplicated.realigned.recalibrated.snpEff.snpSift.vcf

#Dependencies
import sys
import os
import subprocess
from collections import defaultdict

mutations_normal=defaultdict(float)
mutations_t1=defaultdict(float)
mutations_t2=defaultdict(float)
mutations_t3=defaultdict(float)
```

```python
#Read through VCF File
file=open(V4,"r")
skippedindel=0
skippedlowreadcount=0
skippedlowvaf=0
totalsites=0
import sys

for line in file:
    line.strip()
    skipvalue=0
    if line.startswith('#'):
        continue
      # PROCESS LINE HERE
    else:
        chrom,position,ID,reference,variant,quality,ftype,info,formattype,sample=line.split("\t")
        #CHECK 1: is it an indel, if so filter out for now
        refsize=len(reference)
        varsize=len(variant)
        if refsize == varsize:
            try:
                (GT,AD,DP,GQ,PL)=sample.split(":")
                #print("GOOD:",line)
                (refcount,varcount)=AD.split(",")
                vaf=(float(varcount)/(float(refcount)+float(varcount)))*100
                #print(sample,float(vaf),refcount,varcount)
                if (float(vaf) < 5.0):
                    #skippedlowvaf+=1
                    #print("SKIP",str(vaf))
                    skipvalue=1
                    continue
                else:
                    totalsites+=1
                    final=chrom+"\t"+position+"\t"+refcount+"\t"+varcount+"\t"+str(vaf)
                    positionfull=chrom+"_"+position+"_"+reference+"_"+variant
                    #Store this position and associated VAF in dictionary to compare to next sample
                    mutations_t3[positionfull]=vaf
                    #print(final)
            except ValueError:
                #When there are two alt alleles - these are mostly indels
                #BAD: chr2	233345249	rs1190432	C	A,T
                #In this case there are more values for GT,AD,DP,GQ,PL
                #1/2:1,13,5:19:99:515,135,137,347,0,368
                #The second element refers to the first alt allele and the second refers to the second alt allele count
                # 13 - A variant, 5 - T variant, 19 read depth
                #print("BAD:",line)
                skippedindel+=1
                continue
        else:
            skippedindel+=1
            
print(skippedindel,skipvalue)
```

```python
normal=list(mutations_normal)
t1=list(mutations_t1)
t2=list(mutations_t2)
t3=list(mutations_t3)

unique_mutations=set(normal+t1+t2+t3)
```

```python
sys.stdout = open("/Users/rjayasin/Desktop/Mutation_S2_vafs.txt","wt")

for mutation in unique_mutations:
    normal = mutations_normal.get(mutation, "0.0")
    t1 = mutations_t1.get(mutation, "0.0")
    t2 = mutations_t2.get(mutation, "0.0")
    t3 = mutations_t3.get(mutation, "0.0")
    print(mutation,normal,t1,t2,t3)
```
